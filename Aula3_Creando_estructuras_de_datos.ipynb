```python
import pandas as pd
```

# Series

```python
data = [1,2,3,4,5]
```

```python
s = pd.Series(data)
s
```

```python
index = ['Linea' + str(i) for i in range(5)]
index
```

```python
s = pd.Series(data = data, index = index)
s
```

```python
data = {'Linea' + str(i) :i+1 for i in range(5)} #otra forma de hace lo mismo que en las dos lineas de codigo de arriba
data
```

```python
s = pd.Series(data)
s
```

```python
s1 = s +2
s1
```

```python
s2 = s + s1 #se pueden realizar operaciones matemeticas entre series siempre y cuando tengan el mismo indice
s2
```
