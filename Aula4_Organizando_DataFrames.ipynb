```python
import pandas as pd
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
df.sort_index(inplace = True) #ordenando los indices
df
```

```python
df.sort_index(inplace = True, axis = 1)  #ordenando las columnas con axis
df
```

```python
df.sort_values(by = 'X' ,inplace = True)  #ordenando el contenido del DataFrame por los valores de la columna X en forma ascendente
df
```

```python
df.sort_values(by = '3', axis = 1 ,inplace = True)  #ordenando el contenido del DataFrame por los valores de la fila 3 en forma ascendente y ordenacion columnar(?)
df
```
